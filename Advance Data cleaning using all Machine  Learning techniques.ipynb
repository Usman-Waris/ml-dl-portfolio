```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

dataset = pd.read_csv('portugal_house.csv')

numeric_cols = dataset.select_dtypes(include=['number']).columns
dataset[numeric_cols] = dataset[numeric_cols].fillna(dataset[numeric_cols].median())

text_cols = dataset.select_dtypes(include=['object']).columns
for col in text_cols:
    most_frequent = dataset[col].mode()[0]
    dataset[col] = dataset[col].fillna(most_frequent)

dataset = dataset.drop_duplicates()

print("Cleaning complete! NOw our dataset is ready.")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

high_card_cols = [col for col in dataset.select_dtypes(include=['object']).columns
                  if dataset[col].nunique() > 1000]

for col in high_card_cols:
    print(f"Encoding {col} with {dataset[col].nunique()} unique values...")
    dataset[col] = le.fit_transform(dataset[col].astype(str))

print("\nHigh cardinality columns encoded!")
```

```python

remaining_text_cols = dataset.select_dtypes(include=['object']).columns

dataset = pd.get_dummies(dataset, columns=remaining_text_cols, drop_first=True)

print(f"columns after encoding: {dataset.shape[1]}")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = dataset.drop('Price', axis=1)
y = dataset['Price']
X_scaled = scaler.fit_transform(X)

print("congragulations! your 1lak rows dataset is now clean, Encode and Scale .")
```

```python
import numpy as np

print("Mean (should be equal to zero):", np.round(np.mean(X_scaled), 2))
print("Standard Deviation (should be):", np.std(X_scaled))
```

```python

dataset = dataset.astype(int, errors='ignore')

bool_cols = dataset.select_dtypes(include=['bool']).columns
dataset[bool_cols] = dataset[bool_cols].astype(int)

print("All True/False are now 1 and 0!")
```
